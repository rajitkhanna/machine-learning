## Manipulating Data

```python
soccer_data["hometeam"].str.lower()
```

```python
soccer_data
```

```python
soccer_data.bbah = soccer_data.bbah.fillna(soccer_data.bbah.mean()) # we need to reassign to see the change
```

```python
soccer_data
```

```python
soccer_data.bbahh.fillna(soccer_data.bbahh.mean(), inplace=True)
```

```python
soccer_data
```

```python
soccer_data.dropna(inplace=True)
```

```python
# Column from series -> can add partial data
team_column = pd.Series([True] * len(soccer_data))

# New column called team
soccer_data["Team"] = team_column
soccer_data
```

```python
# Column from Python list -> must be same length as existing dataframe
team_location = ["England"] * len(soccer_data)

# New column called location
soccer_data["Location"] = team_location
soccer_data
```

```python
soccer_data["Goal difference"] = soccer_data.fthg - soccer_data.ftag
soccer_data
```

```python
soccer_data.groupby(["hometeam"]).sum()["Goal difference"]
```

```python
soccer_data = soccer_data.drop("Team", axis=1)
soccer_data
```

```python
soccer_data["Team"] = True
soccer_data
```
