# Introduction to Matplotlib

```python
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot(pd.Series([1, 2, 3, 4, 5]), 'b+') # takes y values from pandas dataframe and uses index array [0, 1, ... N] for X
plt.plot(pd.Series([5, 4, 3, 2, 1]), 'r+')
plt.show()
```

```python
plt.plot()
```

```python
plt.plot(); # get rid of empty brackets - could also use plt.show()
```

```python
# pyplot API vs object-oriented API
# in general try to use object-oriented interface over pyplot interface

# Method 1
figure = plt.figure() # create a figure
axes = figure.add_subplot() # adds axes to the figure
plt.show()
```

```python
# Method 2
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1]) # set dimensions of the plot
ax.plot(pd.Series([1, 2, 3, 4, 5]))
plt.show()
```

```python
# 3rd method (recommended)

fig, ax = plt.subplots()
ax.plot(pd.Series([1, 2, 3, 4, 5]));
```
